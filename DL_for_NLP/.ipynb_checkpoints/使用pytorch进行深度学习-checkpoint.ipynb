```python
import torch 
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
# 1准备数据
data = [("me gusta comer en la cafeteria".split(), "SPANISH"),
        ("Give it to me".split(), "ENGLISH"),
        ("No creo que sea una buena idea".split(), "SPANISH"),
        ("No it is not a good idea to get lost at sea".split(), "ENGLISH")]

test_data = [("Yo creo que si".split(), "SPANISH"),
             ("it is lost on me".split(), "ENGLISH")]

# 2建立word2index
word2index={}
data_total=data+test_data
for sentence,_ in data_total:
    for word in sentence:
        if word not in word2index:
            word2index[word]=len(word2index)
           

# 3建立bow
def bow(sample):
    vector=torch.zeros(len(word2index))
    sentence=sample[0]
    if sample[1]=='SPANISH':
        label=torch.LongTensor([1])
    else:
        label=torch.LongTensor([0])

    for word in sentence:
        vector[word2index[word]]+=1
    return vector,label






        
        
        

            
        
        
```

```python
for sample in test_data:
    vector,label=bow(sample)
    print(vector)
    print(label)
```

```python
#建立神经网络
# 1初始化与网络结构设计
# 2 forward

class bowClassify(nn.Module):   #module:模块，组件
    def __init__(self):
        super(bowClassify,self).__init__()
        self.linear=nn.Linear(len(word2index),2)   #nn.后面一定注意是大写，nn.Linear()
        
    def forward(self,vector):
        x=F.log_softmax(self.linear(vector),dim=0)
        x=x.reshape(1,-1)
        return x
        
            


model=bowClassify()



```

```python
vector,label=bow(data[0])
prob=model(vector)  #这里，不用调用model.forward函数，直接使用model()就可以，这是net相对于普通python类不相同的地方
print(prob)
print(label)
```

```python
# 开始训练
def train(model):
    for epoch in range(100):
        loss_function = nn.NLLLoss()  #python函数式编程，给nn.NLLLoss定了一个函数指针loss_function，以后loss_function就可以调用对应函数
        optimizer = optim.SGD(model.parameters(), lr=0.1)
        for sample in data:
            model.zero_grad()
            vector,label=bow(sample)       
            prob=model(vector)
            loss=loss_function(prob,label)
            loss.backward()
            optimizer.step()
        with torch.no_grad():
            
            accuracy=0
            for sample in test_data:
                vector,label=bow(sample)
                prob=model(vector)
                if prob[0][0]<prob[0][1] and label==1:
                    accuracy+=1
                if prob[0][0]>prob[0][1] and label==0:
                    accuracy+=1
            accuracy=accuracy/len(test_data)
            print("epoch{}:accuracy_testdata={}".format(epoch,accuracy))
        
        if epoch==99:
            print("the last training,the parmeters are:")
            for param in model.parameters():
                print(param)
            

train(model)
#这里，说明在train函数中训练，model参数就会参与训练，不用在主程序中训练，这里传进来的参数model是可变变量。


        
```
